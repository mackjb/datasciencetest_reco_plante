```python
from pathlib import Path
import random
from PIL import Image
import matplotlib.pyplot as plt

# chemins de base (à adapter si besoin)
PROJECT_ROOT = Path.cwd()  # si tu lances le notebook à la racine du repo
COLOR_ROOT = PROJECT_ROOT / "dataset/plantvillage/data/plantvillage dataset/color"
OUTPUT_PATH = PROJECT_ROOT / "Streamlit/assets/dataset_overview_color.png"

COLOR_ROOT, OUTPUT_PATH
```

```python
# Lister quelques sous-dossiers (classes)
class_dirs = [d for d in COLOR_ROOT.iterdir() if d.is_dir()]
len(class_dirs), class_dirs[:5]
```

```python
# Sélectionner par exemple 4 classes au hasard
random.seed(42)
selected_dirs = random.sample(class_dirs, k=4)

images = []
for d in selected_dirs:
    # prendre jusqu'à 4 images par classe
    files = list(d.glob("*.jpg")) + list(d.glob("*.JPG")) + list(d.glob("*.png"))
    files = files[:4]
    for f in files:
        images.append(f)

len(images), images[:8]
```

```python
# Garder au plus 16 images
images = images[:16]

n_images = len(images)
n_cols = 4
n_rows = (n_images + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 6))

for idx, ax in enumerate(axes.flatten()):
    if idx < n_images:
        img = Image.open(images[idx])
        ax.imshow(img)
        ax.axis("off")
    else:
        ax.axis("off")

plt.tight_layout()

# Sauvegarde dans Streamlit/assets
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
fig.savefig(OUTPUT_PATH, dpi=150)
plt.close(fig)

OUTPUT_PATH
```
